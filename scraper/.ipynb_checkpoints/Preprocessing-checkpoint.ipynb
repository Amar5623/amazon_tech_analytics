```python
import pandas as pd
from sqlalchemy import create_engine

# Define your connection details
username = 'root'
password = 'root'
host = 'localhost'
port = '3306'
database = 'amazon_scraper'

# Create the SQLAlchemy engine
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

# Load the `products` table into a DataFrame
products_df = pd.read_sql('SELECT * FROM products', con=engine)

# Preview the data
print("✅ Data loaded from MySQL:")
display(products_df.head())
```

```python
# Display the first 5 rows
print("Sample data:")
display(products_df.head())

# Display basic info
print("\nData types and nulls:")
products_df.info()

# Check for missing values
print("\nMissing values per column:")
print(products_df.isnull().sum())
```

```python
import re

# 1. Clean ASIN
products_df['asin'] = products_df['asin'].str.strip()
asin_duplicates = products_df['asin'].duplicated().sum()
print(f" Duplicate ASINs found: {asin_duplicates}")

# Drop rows with duplicate or null ASINs
products_df.drop_duplicates(subset='asin', inplace=True)
products_df.dropna(subset=['asin'], inplace=True)

# 2. Validate URL (must start with http and contain 'amazon')
products_df['url_valid'] = products_df['url'].apply(lambda x: x.startswith('http') and 'amazon' in x)
invalid_urls = products_df[~products_df['url_valid']]
print(f" Invalid URLs found: {len(invalid_urls)}")

# Drop invalid URLs (optional)
products_df = products_df[products_df['url_valid']].drop(columns='url_valid')

# 3. Clean and normalize title
products_df['title'] = products_df['title'].fillna('').str.lower().str.strip()
products_df['title'] = products_df['title'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s\-.,]', '', x))  # Remove unwanted chars

# Preview cleaned columns
print("\n Cleaned columns preview:")
display(products_df[['asin', 'url', 'title']].head())
```

```python
# 1. Clean price (already float, just round for consistency)
products_df['price'] = pd.to_numeric(products_df['price'], errors='coerce').round(2)

# 2. Clean discount – remove '%' and convert to float
products_df['discount'] = products_df['discount'].str.replace('%', '', regex=False)
products_df['discount'] = pd.to_numeric(products_df['discount'], errors='coerce').round(2)

# 3. Clean rating – filter values between 0 and 5
products_df['rating'] = pd.to_numeric(products_df['rating'], errors='coerce')
products_df.loc[(products_df['rating'] > 5) | (products_df['rating'] < 0), 'rating'] = None

# 4. Clean review_count – force to integer, allow NaNs
products_df['review_count'] = pd.to_numeric(products_df['review_count'], errors='coerce').dropna()
products_df['review_count'] = products_df['review_count'].astype('Int64')  # nullable integer type

# 5. Clean brand – normalize text, do not fill missing
products_df['brand'] = products_df['brand'].str.strip().str.lower()

# Show summary of cleaned columns
print("Summary after cleaning:")
print(products_df[['price', 'discount', 'rating', 'review_count', 'brand']].describe(include='all'))

# Optional: Check how many rows have missing values now (just info, no deletion)
print("\nMissing value counts (post-cleaning):")
print(products_df[['price', 'discount', 'rating', 'review_count', 'brand']].isnull().sum())
```

```python
# 1. Normalize category
products_df['category'] = products_df['category'].str.strip().str.lower()

# Optional: Inspect unique categories
print("Unique categories:")
print(products_df['category'].value_counts())

# 2. Convert created_at and updated_at to datetime if not already
products_df['created_at'] = pd.to_datetime(products_df['created_at'], errors='coerce')
products_df['updated_at'] = pd.to_datetime(products_df['updated_at'], errors='coerce')

# 3. Check for invalid or future timestamps
from datetime import datetime

now = pd.Timestamp.now()
future_created = products_df[products_df['created_at'] > now]
future_updated = products_df[products_df['updated_at'] > now]

print(f"\n Future 'created_at' rows: {len(future_created)}")
print(f"Future 'updated_at' rows: {len(future_updated)}")

# Preview time differences if needed
products_df['time_diff'] = (products_df['updated_at'] - products_df['created_at']).dt.total_seconds() / 3600

print("\n Time difference (in hours) between update and creation — stats:")
print(products_df['time_diff'].describe())
```

```python
specs_df = pd.read_sql('SELECT * FROM specifications', con=engine)
```

```python
# Display sample
print("Sample specifications data:")
display(specs_df.head())

# Show basic info
print("\nData types and nulls:")
specs_df.info()

# Missing values
print("\nMissing values per column:")
print(specs_df.isnull().sum())

# Clean spec_name and spec_value
specs_df['spec_name'] = specs_df['spec_name'].astype(str).str.strip().str.lower()
specs_df['spec_value'] = specs_df['spec_value'].astype(str).str.strip()

# Drop rows with missing product_id or spec_name
specs_df = specs_df.dropna(subset=['product_id', 'spec_name'])

# Check for duplicates
duplicates = specs_df.duplicated(subset=['product_id', 'spec_name']).sum()
print(f"\nDuplicate (product_id, spec_name) pairs: {duplicates}")

# Optional: Drop duplicates if needed
# specs_df = specs_df.drop_duplicates(subset=['product_id', 'spec_name'])

# Preview cleaned specs
print("\nCleaned specifications preview:")
display(specs_df.head())
```

```python
# Create a dict to store category-wise pivoted specs
category_specs = {}

# Get distinct categories
categories = products_df['category'].dropna().unique()

# For each category, pivot specs only for products of that category
for cat in categories:
    cat_products = products_df[products_df['category'] == cat]
    cat_specs = specs_df[specs_df['product_id'].isin(cat_products['id'])]

    # Pivot
    pivot = cat_specs.pivot_table(
        index='product_id',
        columns='spec_name',
        values='spec_value',
        aggfunc='first'
    ).reset_index()

    # Clean column names
    pivot.columns.name = None
    pivot.columns = [col.strip().replace(" ", "_") for col in pivot.columns]
    
    # Merge with category-specific product data
    merged = cat_products.merge(pivot, left_on='id', right_on='product_id', how='left')
    
    category_specs[cat] = merged  # Store

    print(f"\n Processed category: {cat}")
    print(f"Columns: {list(merged.columns)}")
    display(merged.head(2))
```

```python
images_df = pd.read_sql('SELECT * FROM images', con=engine)
```

```python
# Display sample data
print(" Sample images data:")
display(images_df.head())

# Info & nulls
print("\n Data types and missing values:")
images_df.info()
print("\n Missing values per column:")
print(images_df.isnull().sum())

# Clean: Strip whitespace and lowercase URLs (optional)
images_df['image_url'] = images_df['image_url'].astype(str).str.strip()

# Remove completely missing URLs or product_id
images_df = images_df.dropna(subset=['product_id', 'image_url'])

# Ensure position is integer
images_df['position'] = pd.to_numeric(images_df['position'], errors='coerce').fillna(0).astype(int)

# Check for duplicates
duplicate_images = images_df.duplicated(subset=['product_id', 'image_url', 'position']).sum()
print(f"\n Duplicate (product_id, image_url, position): {duplicate_images}")
```

```python
# Filter for position = 1
primary_images = images_df[images_df['position'] == 1][['product_id', 'image_url']]

# Rename for clarity
primary_images = primary_images.rename(columns={'image_url': 'primary_image'})

# Join to products_df
products_with_images = products_df.merge(primary_images, left_on='id', right_on='product_id', how='left')

print("\n Products with primary image:")
display(products_with_images[['title', 'primary_image']].head())
```

```python
from sqlalchemy import create_engine, Column, Integer, String, Float, DateTime
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker

# Define the base class for the models
Base = declarative_base()

# Define Product model
class Product(Base):
    __tablename__ = 'products'
    
    id = Column(Integer, primary_key=True, autoincrement=True)
    asin = Column(String(50), unique=True, nullable=False)
    title = Column(String(255), nullable=False)
    price = Column(Float, nullable=True)
    discount = Column(Float, nullable=True)
    rating = Column(Float, nullable=True)
    review_count = Column(Integer, nullable=True)
    brand = Column(String(100), nullable=True)
    category = Column(String(100), nullable=True)
    created_at = Column(DateTime, nullable=True)
    updated_at = Column(DateTime, nullable=True)
    primary_image = Column(String(255), nullable=True)

# Define Specification model
class Specification(Base):
    __tablename__ = 'specifications'
    
    id = Column(Integer, primary_key=True, autoincrement=True)
    product_id = Column(Integer, nullable=False)
    spec_name = Column(String(100), nullable=False)
    spec_value = Column(String(255), nullable=True)

# Define Image model
class Image(Base):
    __tablename__ = 'images'
    
    id = Column(Integer, primary_key=True, autoincrement=True)
    product_id = Column(Integer, nullable=False)
    image_url = Column(String(255), nullable=False)
    position = Column(Integer, nullable=False)

# Create the engine for the processed database
processed_db_url = "mysql+pymysql://root:root@localhost:3306/amazon_scraper_processed"
engine = create_engine(processed_db_url)

# Create the tables in the new database (if they don't already exist)
Base.metadata.create_all(engine)

# Create a session for adding data
SessionLocal = sessionmaker(autocommit=False, autoflush=False, bind=engine)
session = SessionLocal()

# The rest of your code for saving data remains unchanged
# Save processed products data
for index, row in products_df.iterrows():
    product = Product(
        asin=row['asin'],
        title=row['title'],
        price=row['price'],
        discount=row['discount'],
        rating=row['rating'],
        review_count=row['review_count'],
        brand=row['brand'],
        category=row['category'],
        created_at=row['created_at'],
        updated_at=row['updated_at'],
        primary_image=row.get('primary_image', None)  # Handle missing primary image
    )
    session.add(product)

# Commit the changes to the products table
session.commit()

# Save processed specifications data
for index, row in specs_df.iterrows():
    specification = Specification(
        product_id=row['product_id'],
        spec_name=row['spec_name'],
        spec_value=row['spec_value']
    )
    session.add(specification)

# Commit the changes to the specifications table
session.commit()

# Save processed images data
for index, row in images_df.iterrows():
    image = Image(
        product_id=row['product_id'],
        image_url=row['image_url'],
        position=row['position']
    )
    session.add(image)

# Commit the changes to the images table
session.commit()

# Close the session
session.close()

print("✅ Processed data successfully saved to the 'amazon_scraper_processed' database.")
```
